### The accept-reject sampling method

- accept-reject sampling method

 forward sampling에 evidence variable을 포함할 수 있도록 해준다.
 
증거가 나타난다면

1. 표본을 추출하고

2. 증거에 맞는지 확인한다.

3. 맞다면 허락하고 아니면 거절한다.

---

예) 유방암의 확률은 10만명 중 125명이다.

10만명의 표본을 뽑으면 99.875%를 거절한다.

20-30살, 아시아 조상, 남자와 같은 조건을 추가하면 이러한 표본값을 찾기는 거의 불가능해진다. 

그러므로 이 방법은 매우 저차원에서만 가능하다.

### Gibbs sampling

- Gibbs sampling

그래피컬 모델로 부터 표본을 추출하는 MCMC의 한 방법이다. 

이 방법은 각 변수의 조건부 분포가 알려져있거나 표본추출하기 쉬울 때 사용된다. 

Gibbs distribution: 베이즈 네트워크나 마코브 네트워크로 부터 나온 factors의 곱

---

- Steps in the Gibbs sampling procedure

1) 현재 상태로 시작한다.

• Start with the current assignment to all variables

2) 변수의 순서대로 표본을 추출한다.(다른 변수들은 전 상태 고정)

• For j = 1 to number of variables, sample each X j variable, while the other variables hold their previous values

![Image](figures/4.png)

x1,x2,x3는 이항변수, 현재상태는 0,0,0

한번씩 샘플링 했을때 결과값은 x'

요약하면, 깁스 체인은 마코브 체인의 가장 간단한 그래피컬 모델이다. 깁스 체인은 정체 분포에 도달하지 않을 수도 있지만 특정 조건 하에서 섞임을 보장한다. 마코브체인의 가장 간단한 모델이지만 섞이는데 오래걸린다. 또한 factor의 곱으로 부터 추출할 수 있을 때만 사용된다. 

- An example of Gibbs sampling

job interview 예시 사용(2X2X3X2X2=48조합분포)

P(Offer,Grades,Interview| Admission,Experience)에 관심

Admission와 Experience의 관측값 있음



```python
tcpd,bn,skel=getTableCPD()
query={'Offer':'0','Grades':'0','Interview':'0'}
evidence={'Admission':'0','Experience':'0'}
fac=tcpd.condprobve(query,evidence)
df=printdist(fac,bn)
df
```

만약 정확한 추정exact inference(Variable Elimination)을 한다면 결과값은 위와 같다. 

P(Offer,Grades, Interview | Admission = 0, Experience = 0)을 얻기 위해서 표본을 추출하고 증거에 맞지 않는 것을 거절해야 한다.

Libpgm을 통하여 무작위 샘플링과 깁스샘플링을 할 수 있다. 각각 5000번씩 시행하여 이를 비교해보자.

```python
def estimate_distrib(skel,samples):
    learner=PGMLearner()
    #learn the parameters of the network from the samples, given skeleton
    #returns a new bayes net.
    bayesnet=learner.discrete_mle_estimateparams(skel,samples)
    tablecpd=TableCPDFactorization(bayesnet)
    #run a conditional probability query for
    #P(Offer,Grades,Interview| Admission=0,Experience=0)
    fac=tablecpd.condprobve(query,evidence)
    #create a dataframe listing the marginals
    df2=printdist(fac,bayesnet)
    return df2
#learn the marginals from gibbs samples
def gibbs_marginals(num_samples=5000):
    tcpd,bn,skel=getTableCPD()
    samples=tcpd.gibbssample(evidence,num_samples)
    df2=estimate_distrib(skel,samples)
    return df2['probability']
    #learn the marginals from random samples
def random_sample_marginals(num_samples=5000):
    tcpd,bn,skel=getTableCPD()
    samples=bn.randomsample(num_samples,evidence)
    df2=estimate_distrib(skel,samples)
    return df2['probability']
df['prob from gibbs']=gibbs_marginals()
df['prob from random samples']=random_sample_marginals()
df
```

깁스 샘플링이 무작위추출보다 더 효율적인 방법임을 알 수 있다.
